Rozk≈Çad normalny $ X ~ N(\mi,\simga^2)$

```python
Zadanie nr 1  n=1000, simga=2, mi = 2
```

```python
import numpy as np
normal_probes = np.random.normal(loc=2,scale=2,size=1000)
```

```python
import matplotlib.pyplot as plt
from scipy.stats import norm
plt.hist(normal_probes,bins=30, density=True)
x_range = np.arange(-10,10,0.01)
y_range = norm.pdf(x_range, 2,2)
```

```python
import seaborn as sns
sns.ecdfplot(normal_probes)
y_range_ppf = norm.ppf(x_range,2,2)
plt.plot(y_range_ppf,x_range)
```
