```python
import numpy as np 
```

```python
n =1000
B_0=3
B_1=2
sigma=1
alfa = 0.05
x_arr = np.linspace(0,10,1000)
eps = np.random.normal(0,sigma,n)
```

```python
print(eps)
```

```python
y_arr = x_arr*B_1 + B_0 + eps
```

```python
x_mean = np.mean(x_arr[:990])
y_mean = np.mean(y_arr[:990])
```

```python
a_upper = sum([ (x-x_mean)*y for (x,y) in zip(x_arr[:990],y_arr[:990])])
a_down = sum([pow(x - x_mean,2) for x in x_arr[:990]])

b1_est = a_upper/a_down

b0_est = y_mean - x_mean*b1_est

print(b1_est)
print(b0_est)
```

```python
from scipy import stats
prediction_y_arr = []
range_known_sigma = []
range_unknown_sigma = []
for x in x_arr[990:]:
    prediction_y = b0_est + b1_est*x
    prediction_y_arr.append(prediction_y)
    quant = stats.norm.ppf(1-alfa/2, loc=0, scale=1)
    range_known_sigma.append([
        prediction_y - quant*sigma*np.sqrt(1+1/990 + pow((x-x_mean),2)/sum([pow(x_i - x_mean,2) for x_i in x_arr[:990]])),
        prediction_y + quant*sigma*np.sqrt(1+1/990 + pow((x-x_mean),2)/sum([pow(x_i - x_mean,2) for x_i in x_arr[:990]]))
    ])
    quant_gt = stats.t(df=988).ppf((1-alfa/2))
    print(quant_gt)
    s = np.sqrt(sum(pow(prediction_y - y_i, 2) for y_i in y_arr[:990])/(988))
    range_unknown_sigma.append([
        prediction_y - quant_gt*s*np.sqrt(1+1/990 + pow((x-x_mean),2)/sum([pow(x_i - x_mean,2) for x_i in x_arr[:990]])),
        prediction_y + quant_gt*s*np.sqrt(1+1/990 + pow((x-x_mean),2)/sum([pow(x_i - x_mean,2) for x_i in x_arr[:990]]))
    ])
```

```python
print(prediction_y_arr)
```

```python
print(range_known_sigma)
```

```python
import matplotlib.pyplot as plt
```

```python

plt.scatter(x_arr[990:],prediction_y_arr)
plt.scatter(x_arr[990:],y_arr[990:], color='red')
plt.plot(x_arr[990:],range_known_sigma)
```
