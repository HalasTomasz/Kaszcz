```python
n=1000
table_simga_MNW = []
table_sigma_MM = []
for i in range(1000):
    X = np.random.lognormal(1, 0.5, 1000)
    
    ## Metoda najwięszek Wiarygodnośći
    u_MNW = 1/n * sum([np.log(x) for x in X])
    sigma_MNW = 1/n* sum([pow((np.log(x) - u_MNW),2) for x in X])
    table_simga_MNW.append(sigma_MNW)
    
    ### Metdoa momentów
    M1 = 1/n*sum(X)
    M2 = 1/n*sum([pow(x,2) for x in X])
    sigma_MM = np.log(M2) - 2*np.log(M1)  ## NW ?
    table_sigma_MM.append(sigma_MM)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.DataFrame({'Sigma^2 MNW':table_simga_MNW,
                   'Sigma^2 MM':table_sigma_MM})
box_plot  = sns.boxplot(
    data=df,
    showmeans=True,
)
```

```python
import pandas as pd
import numpy as np
with open(r"../Dane/zad1_lista1.txt") as file :
    df_array_1 = []
    df_array_2 = []
    for line in file:
        no_space_line = " ".join(line.split())
        no_space_line = no_space_line.split(" ")
        df_array_1.append(np.float64(no_space_line[0]))
        df_array_2.append(np.float64(no_space_line[1]))


df = pd.DataFrame(data={'x': df_array_1, 'y': df_array_2})
#df = pd.read_csv(r"../Dane/zad1_lista1.txt", sep="   ", header=None,engine='python')
#df = df.apply(lambda row :float(row), axis = 1)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.scatterplot(df['x'],df['y'])
lin_range = np.linspace(0,7,(1000))
rate = np.polyfit(df['x'],df['y'],2)
y_curve = []
print(rate)
for x in lin_range:
    y_curve.append(pow(x,2)*rate[0] + pow(x,1)*rate[1]+rate[0])

sns.lineplot(lin_range,y_curve, label='a*X^2 + b*X^1 + c')
plt.title("wykres rozproszenia")
plt.legend()
plt.savefig('tmp.png')
```

```python
with open(r"../Dane/zad2_lista1.txt") as file :
    df_array = []
    for line in file:
        df_array.append(np.float64(line))
```

```python
plt.scatter(range(len(df_array)),df_array)
```

```python
def pMA(array, p):
    tmp = []
    for i in range(p-2 , len(array)-p+1):
        tmp.append(np.mean(array[i-p:i+p]))
    return tmp
```

```python
time_series = pMA(df_array,101)
```

```python
plt.scatter(range(len(df_array)),df_array)
plt.scatter(range(101,len(time_series)-1),time_series)
plt.legend()
plt.show()
```
