```python
n=1000
table_simga_MNW = []
table_sigma_MM = []
for i in range(1000):
    X = np.random.lognormal(1, 0.25, 1000)
    
    ## Metoda najwięszek Wiarygodnośći
    u_MNW = 1/n * sum([np.log(x) for x in X])
    sigma_MNW = 1/n* sum([pow((np.log(x) - u_MNW),2) for x in X])
    table_simga_MNW.append(sigma_MNW)
    
    ### Metdoa momentów
    M1 = 1/n*sum(X)
    M2 = 1/n*sum([pow(x,2) for x in X])
    sigma_MM = np.log(M2) - 2*np.log(M1) + np.log(n) ## NW ?
    table_sigma_MM.append(sigma_MM)
```

```python
import matplotlib.pyplot as plt

plt.boxplot(table_simga_MNW)
```

```python
plt.boxplot(table_sigma_MM)
```
