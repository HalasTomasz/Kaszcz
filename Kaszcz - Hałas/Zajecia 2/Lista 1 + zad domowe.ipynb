```python
n=1000
table_simga_MNW = []
table_sigma_MM = []
for i in range(1000):
    X = np.random.lognormal(1, 0.5, 1000)
    
    ## Metoda najwięszek Wiarygodnośći
    u_MNW = 1/n * sum([np.log(x) for x in X])
    sigma_MNW = 1/n* sum([pow((np.log(x) - u_MNW),2) for x in X])
    table_simga_MNW.append(sigma_MNW)
    
    ### Metdoa momentów
    M1 = 1/n*sum(X)
    M2 = 1/n*sum([pow(x,2) for x in X])
    sigma_MM = np.log(M2) - 2*np.log(M1)  ## NW ?
    table_sigma_MM.append(sigma_MM)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.DataFrame({'Sigma^2 MNW':table_simga_MNW,
                   'Sigma^2 MM':table_sigma_MM})
box_plot  = sns.boxplot(
    data=df,
    showmeans=True,
)
```

```python
def pMA(array, p):
    tmp = []
    for i in range(p , len(array)-p):
        tmp.append(np.mean(array[i-p:i+p]))
    return tmp
```

```python
time_series = pMA(df_array,11)
```

```python
plt.scatter(range(len(time_series)),time_series)
```
