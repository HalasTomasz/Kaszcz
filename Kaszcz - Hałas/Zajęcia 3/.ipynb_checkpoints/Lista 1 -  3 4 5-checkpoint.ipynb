```python
def pMA(array, p):
    tmp = []
    
    for i in range(p//2, len(array)-p//2):
        #print("indeks: ", i)
        #tmp2 = sum(array[i-p:i+p +1])
        #print(len(range(i-p,i+p)))
        tmp.append(sum(array[i-p//2:i+p//2 +1])/(1+p))
    return tmp
```

### zad 4 

```python
import pandas as pd
import numpy as np
with open(r"../Dane/zad4_lista1.txt") as file :
    df_array_1 = []
    df_array_2 = []
    for line in file:
        no_space_line = " ".join(line.split())
        no_space_line = no_space_line.split(" ")
        df_array_1.append(np.float64(no_space_line[0]))
        df_array_2.append(np.float64(no_space_line[1]))


df = pd.DataFrame(data={'x': df_array_1, 'y': df_array_2})
```

```python
df.head()
```

```python
calculate_mean_x = np.mean(df['x'])
calculate_mean_y = np.mean(df['y'])

a_upper = sum([ (x-calculate_mean_x)*y for (x,y) in zip(df['x'],df['y'])])
a_down = sum([pow(x - calculate_mean_x,2) for x in df['x']])

a = a_upper/a_down
a
b = calculate_mean_y - calculate_mean_x*a
b

y_points = []

for (x,y) in zip(df['x'],df['y']):
    y_points.append(x*a + b)
```

```python
plt.scatter(df['x'],df['y'],alpha=0.5)
plt.plot(df['x'], y_points, color = 'red', label='Regres plot')
plt.legend()
```

```python
error_array = []
for y_i, y_i_prim in zip(df['y'],y_points):
    error_array.append(y_i-y_i_prim)
plt.boxplot(error_array)    
```

```python
new_x_array = []
new_y_array = []
for (x,y) in zip(df['x'], df['y']):
    if y > 125 or y < -125:
        continue
    new_x_array.append(x)
    new_y_array.append(y)

calculate_mean_x = np.mean(new_x_array)
calculate_mean_y = np.mean(new_y_array)

a_upper = sum([ (x-calculate_mean_x)*y for (x,y) in zip(new_x_array,new_y_array)])
a_down = sum([pow(x - calculate_mean_x,2) for x in new_x_array])

a = a_upper/a_down
a
b = calculate_mean_y - calculate_mean_x*a
b

y_points = []

for x in new_x_array:
    y_points.append(x*a + b)
```

```python
plt.scatter(new_x_array,new_y_array,alpha=0.5)
plt.plot(new_x_array, y_points, color = 'red', label='Regres plot')
plt.legend()
```
